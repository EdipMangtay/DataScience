```python
#Merge
```

```python
df_merge1 = pd.read_csv('7-merge_data1.csv')
```

```python
df_merge2 = pd.read_csv('7-merge_data2.csv')
```

```python
df_merge1
```

```python
df_merge2
```

```python
#iki tabloda id bazıları kesişiyor bazıları kesişmiyor 
```

```python
#merge-outer join
```

```python
df_merged_outer = pd.merge(df_merge1,df_merge2,on='Employee_ID',how='outer') #her şeyi kapısoy riki tabloyu birleştiriyor fakat boş olanları da nan olarak almaktadır
```

```python
df_merged_outer
```

```python
#Merge-left-join
```

```python
df_merge_left = pd.merge(df_merge1,df_merge2,on='Employee_ID',how='left')
```

```python
df_merge_left # birinci tabloyu baz alıyor ikinci tabloda eğer birincilerle ilgili bilgi varsa alıyor yoksa almyıor mesela 11-12-13 ü almıyor biirnci dataframe ana olarak kabul ediyor eşleşenleri alıyor 
```

```python
#merge-right join
```

```python
df_merge_right = pd.merge(df_merge1,df_merge2,on='Employee_ID', how='right')
```

```python
df_merge_right # Sağdakileri ana olarak kabul etmektedir
```

```python
#merge-inner-join
```

```python
df_merged = pd.merge(df_merge1,df_merge2,on='Employee_ID',how='inner')
```

```python
df_merged # her iki tabloda da olanı alır 
```
