```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data  = np.linspace(0,10,20)
```

```python
data1 = data*22
```

```python
plt.subplots()
```

```python
my_fig,my_axes = plt.subplots()
```

```python
type(my_fig)
```

```python
my_fig,my_axes = plt.subplots()
my_axes.plot(data,data1, color = "#1E0703",alpha = 0.2)
my_axes.plot(data1,data)
plt.show()
```

```python
(new_fig,new_axes)  = plt.subplots()
new_axes.plot(data1,data1 + 2, color = "blue",linewidth = 0.4)
new_axes.plot(data1,data1 + 10, color = "red",linewidth = 0.5)
new_axes.plot(data1,data1 + 20, color = "red",linestyle = "-.")

new_axes.plot(data1,data1 + 30, color = "#000000",linewidth = 0.5,marker = "o",markersize = 8, markerfacecolor = "red")


plt.show()
```

```python
plt.scatter(data,data1, color="black")
plt.show()
```
