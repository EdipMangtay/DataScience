```python
age_list = [10,20,30,30,30,40,60,70,75]
```

```python
weight_list = [20,60,80,85,86,87,90,95,90]
```

```python
plt.plot(age_list,weight_list,"r")
plt.xlabel("Yaş")
plt.ylabel("Boy")
plt.title("Age vs W")
plt.show()
```

```python
np_age_list = np.array(age_list)
```

```python
np_weight_list = np.array(weight_list)
```

```python
plt.plot(np_age_list,np_weight_list,"g")
plt.xlabel("kilo")
plt.ylabel("yaş")
plt.title("Table")
plt.show()
```
