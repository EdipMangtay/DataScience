```python
import numpy as np
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
age_list = [10,20,30,30,30,40,60,70,75]
```

```python
weight_list = [20,60,80,85,86,87,90,95,90]
```

```python
plt.plot(age_list,weight_list,"r")
plt.xlabel("Yaş")
plt.ylabel("Boy")
plt.title("Age vs W")
plt.show()
```

```python
np_age_list = np.array(age_list)
```

```python
np_weight_list = np.array(weight_list)
```

```python
plt.plot(np_age_list,np_weight_list,"g")
plt.xlabel("kilo")
plt.ylabel("yaş")
plt.title("Table")
plt.show()
```

```python
numpy_arr1 = np.linspace(0,10,20) # 0 dan 10 a kadar 20 tane sayı dön ama araları eşit aralıkta olsun
```

```python
numpy_arr1
```

```python
numpy_arr2 = numpy_arr1**3
```

```python
numpy_arr2
```

```python
plt.plot(numpy_arr1,numpy_arr2)
plt.show()
```

```python
plt.plot(numpy_arr1,numpy_arr2,"b*-") # hem data point hemde çizgiyi verir
plt.show()
```

```python
plt.plot(numpy_arr1,numpy_arr2)
plt.show()
```

```python
plt.plot(numpy_arr2,numpy_arr1)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(numpy_arr1,numpy_arr2,"r-*")
plt.subplot(1,2,2)
plt.plot(numpy_arr2,numpy_arr1,"b--")
plt.show() # iki grafiği yan yana koymak istersem 
```

```python
my_figure = plt.figure()
figure_axes = my_figure.add_axes([0.1,0.1,0.4,0.3])
figure_axes.plot(numpy_arr1,numpy_arr2,"g")
figure_axes.set_xlabel("X Axis")
figure_axes.set_ylabel("Y Label")
figure_axes.set_title("Graph Title")
plt.show()
```

```python
new_fig = plt.figure(dpi=300)
new_axes = new_fig.add_axes([0.1,0.1,0.3,0.3])
new_axes.plot(numpy_arr1,numpy_arr1**2,label = "numpyArray **2")
new_axes.plot(numpy_arr1,numpy_arr1**3,label = "numpyArray **3")
new_axes.legend(loc = 2)
plt.show()
```
